### 3 - Best time to tweet

creating function that asks from python to ignore nan values, and to take every 'time' value and to change it's type to int, in order to create round hour range that this hour belongs to, by adding +1 to it.
asking from the function to return this range using string formatting.

```python
def hour_range(time_created):
  
    if pd.isna(time_created):
        pass
    else:
        x=int(time_created)
        return (f"{x}:00-{(x+1)}:00")
               
```

applying my function on the 'Created At' column, after applying dt method on it in order to get the hour number ( without mintutes and so), and creating a new column that will store the values that the function will return.

```python
df['Hour_Range']= (df['Created_At'].dt.hour).apply(hour_range)
```

creating bubble sort function to sort the array by index, the function takes two consecutive values and check if the right one is bigger then the left one, if it's true the function swipes them, and continues to the next to values. the range is (-1) so we will not go to non existing index, and (-i) beacuse the rightest value will be always sorted because of the way the function compare every two values.
asking the function to return the sorted array.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n - i - 1):
            if int(arr[j].split(":")[0]) > int(arr[j + 1].split(":")[0]):
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

creating two masks, one that filter and show only the dates before elections including the date Trump was elected, and other filter that shows only the dates after elections

```python
mask= df['Created_At']<=DATE_ELECTED
mask1=df['Created_At']>DATE_ELECTED
```

taking the unique hour range values(without reapet) and applying the function on it.
grouping the values by 'Hour Range', and counting the mean values of 'Retweet_Count','Favorite_Count' for the grouped values, using "reindex" and giving the sorted values to it. creatig bar plot for it, before and after election.
also, setting x,y labels, titles and fontsize for them, also style and legeng location.
asking from python to show the plots.

```python
hour_range_unique= bubble_sort(df[mask]['Hour_Range'].unique())


df[mask].groupby('Hour_Range')[['Retweet_Count','Favorite_Count']].mean().reindex(hour_range_unique).plot(kind='bar',figsize=(20, 10), colormap='Set3',grid=True)
plt.ylabel("Engagement", fontsize=14)
plt.xlabel('Hours', fontsize=14)
plt.title('What the best time for Trump to tweet - Before Elections', fontsize=16)
plt.style.use('seaborn-darkgrid')
plt.legend(loc='upper left',fontsize=12)
plt.show()

df[mask1].groupby('Hour_Range')[['Retweet_Count','Favorite_Count']].mean().reindex(hour_range_unique).plot(kind='bar',figsize=(20, 10),grid=True, colormap='Set3')
plt.ylabel("Engagement", fontsize=14)
plt.xlabel('Hours', fontsize=14)
plt.title('What the best time for Trump to tweet - After Elections', fontsize=16)
plt.style.use('seaborn-darkgrid')
plt.legend(loc='upper left',fontsize=12)
plt.show()
```

*We can see that before Trump was elected the best time for him to tweet was 22:00-23:00 ,and after election it was 08:00-09:00. After elections it didn't metter so much as before elections,his tweets were less retweeted and favorited in general, and got almost same amount of that engagement in every hour of the day.*

### 5 - Best tweet length


writing function that takes tweet, and calculates it's lenght, and returns verbal lenght description depents on the lenght  ("Very short", "Short", "Medium", "Long" or "Very long")

```python
def what_lenght(tweet):
    if len(tweet)<70:
        return "Very Short"
    elif len(tweet)>=70 and len(tweet)<110:
        return "Short"
    elif len(tweet)>=110 and len(tweet)<130:
        return "Medium"
    elif len(tweet)>=130 and len(tweet)<140:
        return "Long"
    elif len(tweet)>=140:
        return "Very Long"
```

I applied the function on 'Text' column, while creating 'Tweet Lenght' column that accept all the values the the function returns

```python
df['Tweet_Lenght']= df['Text'].apply(what_lenght)
```

using the masks the I already created to filter dates before and after elections, grouping the values by 'Tweet Lenght', counting mean values of 'Retweet_Count','Favorite_Count'. Using 'reindex' to sort the future bars the way I want,creating horizontal bar plots. setting size,theme color, x/y labeles, titles, font sizes, style, and location for legened.
asking from python to show the plots/

```python
df[mask].groupby('Tweet_Lenght')[['Retweet_Count','Favorite_Count']].mean().reindex(["Very Short", "Short", "Medium", "Long" , "Very Long"]).plot(kind='barh',figsize=(20, 10), colormap='tab20')
plt.xlabel('Tweet_Lenght',fontsize=12)
plt.ylabel('Engagement',fontsize=12)
plt.title('Best tweet length for Trump - Before Elections', fontsize=14)
plt.style.use('seaborn-darkgrid')
plt.legend(loc='lower right')
plt.show()

df[mask1].groupby('Tweet_Lenght')[['Retweet_Count','Favorite_Count']].mean().reindex(["Very Short", "Short", "Medium", "Long" , "Very Long"]).plot(kind='barh',figsize=(20, 10), colormap='tab20')
plt.xlabel('Tweet_Lenght',fontsize=12)
plt.ylabel('Engagement',fontsize=12)
plt.title('Best tweet length for Trump - After Elections', fontsize=14)
plt.style.use('seaborn-darkgrid')
plt.legend(loc='lower right')
plt.show()
```

*We can see that the best lenght for Trump's tweets before the elections is 'Long' (between 130 to 140 letters) and after the elections 'Very Long' (over 140 letters).*

### 6 - Tweet length distribution

creating two arrays, 'before' and 'after' (that representing before and after election), that counting the values of 'Tweet Lenght' of mask and mask1 using 'tolist()'.
creating two variables 'before_counter' and 'after_counter' with the value '-1', to use them in the functions ' 'before_helper_func' and 'after_helper_func'. '-1' because i  want to use them for index, and index starts from 0.
In the functions I'm using 'global' in order to use the variables and change them, adding +1 to the counters, and returning the values that in the before/after list in some index.

using masks of before and after, counting values in the 'Tweet Lenght' column, and creating plots from it. calling the function I've created to "autopct" in order to show in the pie chart the actual values of the total amount of tweets in each length category. 
setting y labels, titles, fint sizes, and legend location.
asking from python to show the plots.

```python
before = df[mask]['Tweet_Lenght'].value_counts(dropna=False).tolist()
after = df[mask1]['Tweet_Lenght'].value_counts(dropna=False).tolist()

before_counter = -1
after_counter = -1

def before_helper_func(val):
    global before_counter   
    before_counter+=1
    return before[before_counter]

def after_helper_func(val):
    global after_counter   
    after_counter+=1
    return after[after_counter]

df[mask]['Tweet_Lenght'].value_counts().plot(kind='pie',figsize=(10, 10), colormap='Pastel2',autopct=before_helper_func)
plt.ylabel('Tweet_Lenght',fontsize=12)
plt.title('Trump'"s Tweet length distribution - Before Elections", fontsize=14)
plt.legend(loc='upper left')
plt.show()

df[mask1]['Tweet_Lenght'].value_counts().plot(kind='pie',figsize=(10, 10), colormap='Pastel2',autopct=after_helper_func)
plt.ylabel('Tweet_Lenght',fontsize=12)
plt.title('Trump'"s Tweet length distribution - After Elections", fontsize=14)
plt.legend(loc='upper left')
plt.show()
```

*We can see that before elections Trump's mostly tweeted 'Long' tweets (between 130 and 140 letters), and after elections he tweeted 'Very Long' tweets (over 140 letters).*
